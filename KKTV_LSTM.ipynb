```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
train_X = pd.read_csv("light_train_source_labels.csv")
train_X = train_X.drop("user_id", axis=1)
train_y = pd.read_csv("light_train_target_labels.csv")
train_y = train_y.drop("user_id", axis=1)

# Convert the data into the appropriate format for the LSTM model
train_X = np.array(train_X)
train_X = train_X.reshape(train_X.shape[0], train_X.shape[1], 1)  # Reshape to (samples, timesteps, features)
```

```python
# Create the LSTM model
model = Sequential()
model.add(LSTM(units=64, input_shape=(train_X.shape[1], 1)))
model.add(Dense(units=28, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy')

# Fit the model on the training data
model.fit(train_X, train_y, epochs=1, batch_size=128)


```
